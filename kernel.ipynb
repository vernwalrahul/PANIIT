```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.image import *
from keras.layers import *
from keras.optimizers import *
from keras.models import Model, Sequential
from keras import backend as K
from keras.layers import concatenate
from keras.engine.topology import Layer
from keras import initializers, regularizers, constraints, optimizers, layers
from keras.layers import Input, Embedding, Dense, Conv2D, MaxPool2D
from keras.layers import Reshape, Flatten, Concatenate, Dropout, SpatialDropout1D
import keras
from keras.callbacks import ModelCheckpoint
from sklearn.metrics import accuracy_score
```

```python
data_file = 'training/training/'
labels_file = 'training/solution.csv'
labels = to_categorical(list(pd.read_csv(labels_file,index_col=False).iloc[:,1]))
print(labels.shape)
data = []
for i in tqdm(range(1,5001)):
  img = img_to_array(load_img(data_file+str(i)+'.png',target_size=(128,128)))
  data.append(img)
data = np.array(data).astype('float32')
data/=255
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.08)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15)
```

```python
batch_size = 32
num_classes = 7
epochs = 100
input_shape = X_train[0].shape
print("input_shape = ", input_shape)
# model = Sequential()
# model.add(Conv2D(64, kernel_size=(3,3),
#                  activation='elu',input_shape=input_shape, kernel_initializer='glorot_uniform'))
# model.add(SpatialDropout2D(0.2))
# model.add(MaxPooling2D(pool_size=(2, 2)))
# model.add(Conv2D(64, (5, 5), activation='elu',kernel_initializer='glorot_uniform'))
# model.add(SpatialDropout2D(0.25))
# model.add(MaxPooling2D(pool_size=(4,4)))
# model.add(Flatten())
# model.add(Dense(32, activation='relu'))
# model.add(Dropout(0.5))
# model.add(BatchNormalization())
# model.add(Dense(num_classes, activation='softmax'))

filter_sizes = [3,3,3,5,5,5]
num_filters = 64

inp = Input(shape=input_shape)

maxpool_pool = []
for i in range(len(filter_sizes)):
    conv = Conv2D(num_filters, kernel_size=(filter_sizes[i],filter_sizes[i]),
                                 kernel_initializer='he_normal', activation='elu',padding='same')(inp)
    maxpool_pool.append(MaxPool2D(pool_size=(2,2))(conv))

z = Concatenate(axis=1)(maxpool_pool)
z = SpatialDropout2D(0.3)(z)
z = Conv2D(64,kernel_size=(5,5),kernel_initializer='he_normal', activation='elu')(z)
z = MaxPool2D(pool_size=(2,2))(z)
z = Flatten()(z)
z = Dense(32, activation="relu")(z)
z = Dropout(0.5)(z)
z = BatchNormalization()(z)
outp = Dense(num_classes, activation="softmax")(z)

model = Model(inputs=inp, outputs=outp)
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer='Adam',
              metrics=['accuracy'])

checkpoint = ModelCheckpoint('model.chk', monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

print(model.summary())
# model.fit(X_train, y_train,
#           batch_size=batch_size,
#           epochs=epochs,
#           verbose=1,
#           callbacks=callbacks_list,
#          validation_data=(X_val,y_val))
```

```python
train_datagen = ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
)

train_datagen.fit(X_train)

# this is the augmentation configuration we will use for testing:
# only rescaling
# test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = train_datagen.flow(
    X_train,y_train,
    batch_size=batch_size)

validation_generator = train_datagen.flow(
    X_val,y_val,
    batch_size=batch_size)

model.fit_generator(
    train_generator,
    steps_per_epoch=5000// batch_size,
    epochs=epochs,
    verbose=1,
    callbacks=callbacks_list,
    validation_data=validation_generator)
```

```python
y_pred = model.predict(X_test).argmax(axis=1)
accuracy_score(y_test.argmax(axis=1),y_pred)
```
